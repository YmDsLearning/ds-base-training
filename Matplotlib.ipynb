# Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## plt.plot()

```python
x=[1, 2, 3, 4, 5]
y=[1, 4, 9, 16, 25]
plt.plot(x, y)
```

```python
#plt.plot()にxを入れないこともできる
#ただし、yのindexが横軸にくる（xではない！）ので注意
plt.plot(y)
```

```python
x=[1, 2, 3, 4, 5]
#yを数式化してみる
#リストに対しては
#y = x ** 2
#という式は使えないので、NumPyを使用する
y = np.array(x) ** 2
y
```

```python
plt.plot(x, y)
```

### シグモイド関数を描画してみる

```python
#シグモイド関数(z)をPythonで書く
#数式のグラフを表示するときは、横軸にはNumPyのlinspaceを使う
x = np.linspace(-10, 10, 10)
#e: np.exp
z = 1/ (1 + np.exp(-x))
plt.plot(x, z)
```

```python
x
```

```python
#DataFrameのplotを描画する
df = pd.DataFrame({
    'm': [1, 2, 3, 4, 5, 6],
    'n': [10, 15, 40, 100, 200, 400],
})
```

```python
df
```

```python
#aが横軸、bが縦軸
#Seriesで渡す場合
plt.plot(df['m'],df['n'])
```

```python
#columnの情報だけを渡すやり方
#data引数にDataFrameを指定する
plt.plot('m', 'n', data=df)
```

```python
#Seriesが１つだけだと、Seriesのindexが横軸に表示されるので注意
plt.plot(df['n'])
```

## 装飾

```python
x = np.linspace(-10, 10, 10)
z = 1/ (1 + np.exp(-x))
plt.plot(x, z)
```

```python
#線の色
#b: 青
#r: 赤
#y: 黄色
#g: 緑
plt.plot(x, z, 'y')
```

```python
#点線
#線の太さ
plt.plot(x, z, 'y--', linewidth=5)
```

```python
#線いろいろ
plt.plot(x, z, 
         color='red', #色
         lw=3,        #太さ
         ls='--',     #形状
         marker='o',  #マーカー
         markersize=10,#マーカーのサイズ
         markerfacecolor='green', #マーカーの中の色
         markeredgecolor='blue',  #マーカーの線の色
         markeredgewidth=3,       #マーカーの線の太さ
         alpha=0.2 #plotの透明度0-1（０に近いほど透明になる）
        )
```

### 付属情報（タイトル、ラベル etc...）をつける

```python
plt.plot(x, z, label='sigmoid')      
plt.plot(x, z*2, label='sigmoid * 2') #複数のグラフを表示する(最後に追加すると凡例がつかないのでここに追加する)
plt.xlabel('x')         #横軸のラベル
plt.ylabel('y')         #縦軸のラベル
plt.title('sigmoid(x)') #タイトル
plt.legend(loc=4)       #凡例(plt.plotのlabelが必要)
plt.xticks([1, 2, 3], rotation=20)   #xの表示したいティックス（rotationは文字の傾き）
plt.yticks([0, 0.5, 1], rotation=20) #yの表示したいティックス（rotationは文字の傾き）
```
