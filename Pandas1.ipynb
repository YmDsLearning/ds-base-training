## Series

```python
#ここでは1レコードをSeriesとして作成
#dictionaryからseriesを作る
data = {
    'Name': 'John',
    'Sex': 'male',
    'Age': 22
}
```

```python
pd.Series(data)
```

```python
#Numpy Arrayから作成 - Age
array = np.array([22, 31, 42, 23])
array
```

```python
#Numpy Arrayから作った場合はindexがない
age_series = pd.Series(array)
age_series
```

```python
#columnの情報を入れたい場合、ラベルをつけることができる
array = np.array(['John', 'male', 22]) #ラベルなし
pd.Series(array)
```

```python
john_series = pd.Series(array, index=['Name', 'Sex', 'Age']) #ラベルをつける
john_series
```

```python
#値を取り出す(dictionaryの使い方と同じ)
john_series['Name']
```

```python
#SeriesはNumPyと同じように統計量の計算ができる
age_series.values
```

```python
age_series.mean()
```

```python
age_series.max()
```

## DataFrame

### pd.DataFrame()

```python
#NumPy Arrayから作成
ndarray = np.arange(10).reshape(2,5)
ndarray
```

```python
pd.DataFrame(ndarray)
```

```python
#項目名をつける: index - 行, columns - 列
#一般的にはindexはそのままでもいいが、columnsにはちゃんと名前をつける
pd.DataFrame(ndarray, index=['index1', 'index2'], columns=['a', 'b', 'c', 'd', 'e'])
```

```python
#dictionaryから作成
#recordが各行になるイメージ
#それぞれのdictionaryが同じ値を持っていること
data1 = {
    'name': 'John',
    'sex': 'male',
    'age': 22
}
data2 = {
    'name': 'Zack',
    'sex': 'male',
    'age': 30
}
data3 = {
    'name': 'Emily',
    'sex': 'female',
    'age': 42
}
#DataFrameにlist形式で入れる
pd.DataFrame([data1, data2, data3])
```

```python
#一番使われる書き方 - 1つのdictionaryから作成
data = {
    'name': ['John', 'Zack', 'Emily'],
    'sex': ['male', 'male', 'female'],
    'age': [22, 30, 42]
}
df = pd.DataFrame(data)
```

```python
#Data Scienceではデータ容量が大きくなりがちなので、１つの変数を使い回してメモリを節約する
df
```

### .drop()

```python
#指定カラム、行の削除が可能
df = pd.read_csv('dataset/tmdb_5000_movies.csv')
```

```python
df.head()
```

```python
#行を削除する
df.drop(0)
```

```python
#カラムを削除する
#axis=1を指定する
df.drop('id', axis=1)
```

```python
#元のDataFrameは変更されない
df
```

```python
#元のDataFrameを変更したい場合
#方法１: inplace=Trueを指定する
df.drop('id', axis=1, inplace=True)
```

```python
df
```

```python
#元のDataFrameを変更したい場合
#方法2: 同じ変数を更新する（こちらの方が一般的）
df = df.drop(0)
```

```python
df
```
