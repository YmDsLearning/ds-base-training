## DataFrameのIteration

### .apply()

```python
#関数を各行に適用(apply)させる
```

```python
df = pd.read_csv('dataset/tmdb_5000_movies.csv')
```

```python
df.head(1)
```

```python
#budgetが０のものはNaNに置き換える
#applyするには、まず関数を作る必要がある
#budgetが０ならNaNを返す
#そうでなければ値をそのまま返す
def return_nan_if_zero(budget):
    if budget == 0:
        return np.nan
    else:
        return budget

return_nan_if_zero(1000)
```

```python
return_nan_if_zero(0)
```

```python
#全ての行にapplyする
#引数にbudgetを取りたいので、columnを指定する->Seriesが取れる
df['budget']
```

```python
#上記に対してapplyする->各budgetに対して関数を実行した値がSeriesで返ってくる
df['budget'].apply(return_nan_if_zero)
```

```python
#元のDataFrameを上書き
#もし存在しないcolumnを指定したら、元のDataFrameに新しく追加される
df['budget'] = df['budget'].apply(return_nan_if_zero)
```

```python
df
```

```python
#1行で書くことができる
def return_nan_if_zero(budget):
    return np.nan if budget == 0 else budget
```

```python
return_nan_if_zero(100)
```

```python
#Lambda関数(1行の関数を無名関数にすることができる)で書くことが多い
x = lambda budget: np.nan if budget == 0 else budget
```

```python
budget = 100
x(budget)
```

```python
#apply()の引数にLambda関数を入れることも可能
df['budget'] = df['budget'].apply(lambda budget: np.nan if budget == 0 else budget)
df['revenue'] = df['revenue'].apply(lambda revenue: np.nan if revenue == 0 else revenue)
df
```

```python
#NaNのSeriesを除外する
df = df.dropna(subset=['budget','revenue'])
```

```python
df
```

```python
#profit = revenue - budget
#lambdaにrowを指定する場合はaxis=1とする
df['profit'] = df.apply(lambda row: row['revenue'] - row['budget'], axis=1)
```

```python
df
```

```python
#profitが最大のレコードを求める
df.iloc[df['profit'].idxmax()]
```

## ピボットテーブル

### .pivot_table()

```python
#ユーザの支払いトランザクションのテーブル
data = {
    'Date': ['Jan-1', 'Jan-1', 'Jan-1', 'Jan-2', 'Jan-2', 'Jan-2'],
    'User': ['Emily', 'John', 'Nick', 'Kevin', 'Emily', 'John'],
    'Method': ['Card', 'Card', 'Cash', 'Card', 'Cash', 'Cash'],
    'Price': [100, 250, 200, 460, 200, 130],
}
df = pd.DataFrame(data)
df
```

```python
#最終的に集計したい値をvaluesに入れ、集計軸をindexとcolumnsに入れる
df.pivot_table(values='Price', index=['Date', 'User'], columns=['Method'])
```

```python
df.pivot_table(values='Price', index=['Date', 'Method'], columns=['User'])
```

```python
#ピボットテーブルから値を取りたい時
#xs(クロスセクション)
#ピボットテーブルの指定の行だけをうまく抜き出す。levelでindexのキーを指定する
#それぞれの行をまたいで値を取るイメージ
pivot_df = df.pivot_table(values='Price', index=['Date', 'Method'], columns=['User'])
pivot_df
```

```python
pivot_df.xs('Card', level='Method')
```

```python
pivot_df.xs('Jan-1')
```

## DataFrameの保存

### .to_csv()

```python
#profit (revenue - budget) の高い順に並べた表を作成する(revenueやbudgetが０のものは除く)
df = pd.read_csv('dataset/tmdb_5000_movies.csv')
df = df[~((df['revenue'] == 0) | (df['budget'] == 0))]
df['profit'] = df.apply(lambda row: row['revenue'] - row['budget'], axis=1)
df = df.sort_values('profit', ascending=False)
```

```python
#CSVファイルに保存
df.to_csv('dataset/tmdb_5000_movies_profit_sorted.csv')
```

```python
#上記のファイルを確認
updated_df = pd.read_csv('dataset/tmdb_5000_movies_profit_sorted.csv')
updated_df
```

```python
#ファイルパスのみ指定で保存した時にはUnnamedというcolumnが追加されている
#元々のindex
#不要の時にはファイル保存の際にindex=Falseを指定する
df.to_csv('dataset/tmdb_5000_movies_profit_sorted.csv', index=False)
```
